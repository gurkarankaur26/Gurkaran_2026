```python
def calbase10log(n):
    if n <1000:
     #print(n)
     pow=3
     while(pow>0):
        #print('power',pow)
        res=int(10**pow)
        #print(res)
        if res==n:
         break
        else:
         pow = pow -.001
        
     print(f"Log10({n}) is {pow}")    
    else:
     print('No. should not be greater than 1000')
   
   
    
```
