```python
def calbase10log(n):
        p = 0
        ctn = True
        step = 0.1
        while ctn:
            if 10 ** (p + step) <= n:
                p = p + step
            else:
                step = step / 10
                if step < 1e-6:
                    ctn = False

        print(f"Log10({n}) is {p:.6f}")  
    
```

```python
calbase10log(231)
```
