```python
import math
def square(num):
    return num*num
def cube(x):
    return x * x * x
def sin(x):
    return math.sin(x)

def approx_derivative(func, x, h):
    #h is the small change
    x1= x
    x2=x+h
    y1 = func(x)
    y2 = func(x+h)
    rate_of_change =  abs((y2-y1))/abs(x2-x1)
    return rate_of_change    
```

```python
approx_derivative(square, 3, 0.0001)
```

```python
import random
```

```python
def print_random_no(start,end):    
    rno =random.random() # generates random number between 0-1
    num= int(rno*end)
    if num < start:
        print(num+start)
    else:
        print(num)
        
```

```python
print_random_no(5,10)
    
```

```python
def toss_coin():
    rno = random.random() #random no between 0 and 1
    mid = 1/2
    if rno < mid:
        print('Head')
    else:
        print('Tail')
    
```

```python
toss_coin()
```

```python
def my_impurity(c1, c2):  
    impurity_value= None   
    if   c1 > 0:    
         impurity_value = c2/c1
    elif c2 >0:
         impurity_value = c1/c2   
    print ('Impurity between '+ str(c1) + ' and '+ str(c2)+': '+str(impurity_value))
```

```python
my_impurity(5,5)
my_impurity(5,4)
my_impurity(5,3)
my_impurity(5,2)
my_impurity(5,1)
my_impurity(5,0)
```
