```python
def sort_list(num):
    len_n = len(num)
    for i in range(len_n):
        for  j in range(len_n-1):
            if num[j] > num[j+1]:
                temp = num[j+1]
                num[j+1] = num[j]
                num[j] =temp
    return num
                
    
```

```python
sort_list([5,45,6,8,3,67,2])
sort_list([0, 1, 0, 0, 1, 0, 1])
```

```python
class Node():
    def __init__(self, num, left=None, right=None):
        self.Key = num
        self.Left = left
        self.Right = right

def insert(new_val, root_node):
    if root_node is None:
        return Node(new_val)

    # Use a temporary pointer to traverse
    current = root_node
    
    while True:
        if new_val < current.Key:
            if current.Left is None:
                current.Left = Node(new_val)
                break
            else:
                current = current.Left
        else: 
            if current.Right is None:
                current.Right = Node(new_val)
                break
            else:
                current = current.Right
    return root_node

def searchbst(val,root_node:Node):
    current = root_node
    while current is not None:            
            if val == current.Key:
               return True
            elif val < current.Key:
                current = current.Left
            else: 
                current = current.Right    
    return False       

def printbst(root_node:Node):
    if root_node is not None:        
            printbst(root_node.Left)
            print(root_node.Key)
            printbst(root_node.Right)
        
```

```python
root = Node(23)
insert(10, root)
insert(11, root)
print(root.Key,root.Left.Key, root.Left.Right.Key)
insert(25, root)
# Result: 23(L:10(R:11), R:25)
print(root.Key,root.Right.Key,root.Left.Key)
insert(45, root)
searchbst(45,root)
printbst(root)
```

```python
def merge (arr1,arr2):
    result=[]
    len_1 = len(arr1)
    len_2 = len(arr2)
    i=0
    j=0
    while i < len_1 and j <len_2:
        if arr1[i] < arr2[j]:
          result.append(arr1[i])
          i+=1          
        elif arr2[j]<arr1[i]:
          result.append(arr2[j])
          j+=1
        else:
            result.append(arr1[i])
            result.append(arr2[j])
            i+=1
            j+=1
    # Add remaining elements of arr1
    while i < len_1:
        result.append(arr1[i])
        i+=1
    # Add remaining elements of arr2
    while j < len_2:
        result.append(arr2[j])
        j+=1
    #print(result)
    return result
```

```python

def merge_sort(arr):
    
    if len(arr) <= 1:
        return arr    
    mid_index = len(arr) // 2
    # Split the list into two halves
    arr1 = arr[:mid_index]
    arr2 = arr[mid_index:]    
    arr1 = merge_sort(arr1)
    arr2 =merge_sort(arr2) 
    #print('before Merge', arr1,arr2)
    return merge(arr1, arr2)
    
```

```python
class DT:
    def __init__(self, boundary=None, boundary_val=None, left=None, right=None,decision =None):
        self.boundary = boundary
        self.boundary_val = boundary_val
        self.left = left 
        self.right = right 
        self.decision =decision
    def YES():
          return DT(decision=True)
    def NO():
          return DT(decision=False)  
    def check(self, features:dict):
        if self.left == None and self.right==None:
            if self.decision == None:
               raise ValueError("Leaf node is missing")
            else:
                return self.decision  
        # Feature validation
        if self.boundary not in features:
            raise ValueError(f"Feature '{self.boundary}' is  missing")
        if features[self.boundary] <= self.boundary_val:
            if self.left:
              return self.left.check(features)
            else:
             raise ValueError("Left node is missing")
        elif features[self.boundary] > self.boundary_val:             
             if self.right:
                return self.right.check(features)
             else:
                raise ValueError("Right node is missing")
        else: 
         return self.decision
            
            
```

```python

nightshift_node = DT(
    boundary ="nightshift",
    boundary_val=0,
    left= DT.YES(),
    right =DT.NO()
)

coffee_node = DT(
    boundary ="coffee",
    boundary_val=2,
    left= DT.NO(),
    right =nightshift_node
)

distance_node = DT(
    boundary ="distance",
    boundary_val=10,
    left= coffee_node,
    right =DT.NO()
)
job_node = DT(
    boundary="salary",
    boundary_val=50000,
    left=DT.NO(),
    right=distance_node    
)
    
print(job_node.check({"salary":10000,"distance":4,"coffee":2,"nightshift":0}))
print(job_node.check({"salary":60000,"distance":4,"coffee":2,"nightshift":0}))
print(job_node.check({"salary":60000,"distance":14,"coffee":2,"nightshift":0}))
print(job_node.check({"salary":60000,"distance":4,"coffee":1,"nightshift":0}))
print(job_node.check({"salary":60000,"distance":4,"coffee":3,"nightshift":1}))
print(job_node.check({"salary":60000,"distance":4,"coffee":3,"nightshift":0}))
```
