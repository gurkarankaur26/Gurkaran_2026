```python
def calculate_hypotenuse(base,height):
    hypotenuse = (base**2 + height**2) ** 0.5
    return hypotenuse
```

```python
calculate_hypotenuse(3, 4)
```

```python
def find_distance_2d(x1,y1,x2,y2):   
    x = x2 - x1    
    y = y2 -y1
    #print(x,y)
    distance = (x**2 +y**2)**0.5
    return distance
```

```python
find_distance_2d(0,0,3,4)
```

```python
def manhattan_distance(x1, y1, x2, y2):
   return  abs(x1-x2)+abs(y1-y2)
```

```python

def predict(m,c,x):
    # y -> point on vertical axis, m -> slope, x -> point on the horizontal axis , 
    #c -> intercept, where the line crosses the y axis
    y = m*x+c
    return y
```

```python
predict(2, 3, 4)  
```

```python
def fit(x1, y1, x2, y2):
    #y = m*x+c
    # y -> point on vertical axis, m -> slope, x -> point on the horizontal axis , 
    #c -> intercept, where the line crosses the y axis
    m = abs((y2-y1)/(x2-x1))
    c = abs(y1- (m*x1))
    return (m,c)
    
```

```python
fit(1, 2, 3, 6)    
```

```python
def square(x):
    return x * x

def approx_derivative(func, x, h):
    func =square
    derivative = (func(x+h)-func(x))/h
    return derivative
    
    
```

```python
def cube(x):
    return x * x * x

def approx_derivative(func, x, h):
    derivative = (func(x+h)-func(x))/h
    return derivative
    
```

```python
print(approx_derivative(cube, 2, 0.0001))
```
